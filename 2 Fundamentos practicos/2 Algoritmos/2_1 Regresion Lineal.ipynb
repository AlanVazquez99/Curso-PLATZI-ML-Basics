```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
df = pd.read_csv('./content/salarios.csv')
df.info()
```

```python
def add_dim(arrays, axis = 1):
    if axis == 0:
        axis = -1

    for array in arrays:
        yield array.reshape(-axis, axis)
```

```python
x = df.iloc[:, 0].values
y = df.iloc[:, 1].values
```

```python
x_train, x_test, y_train, y_test = add_dim( train_test_split(x, y, test_size=0.2, random_state=0), 1 )
```

```python
x_train
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
plt.scatter(x_train, y_train, color='b')
plt.plot(x_train, regressor.predict(x_train), color='black')
plt.title('Salario vs Experiencia')
plt.xlabel('Experiencia')
plt.ylabel('Salario')
```

```python
score = regressor.score(x_test, y_test)
print('Precicion:  {:.2%}'.format(score))
```
