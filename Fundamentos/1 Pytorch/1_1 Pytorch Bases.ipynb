```python
import torch
import numpy as np
```

```python
A = torch.ones(2,3,4,5)
print(A)
print('\n',A.shape)
```

```python
B = torch.Tensor(2,2,2)
print(B)
B.uniform_(-1,1)
print('\n',B)
```

```python
C = torch.rand(2,2,2)
print(C)
```

```python
print(C+B)
print('\n', C*B)
```

```python
print(f'{C}\n{C.shape}\n')
reshaped = C.view(4,2)
print(f'{reshaped}\n{reshaped.shape}')
```

```python
points = torch.tensor([[1.,2.], [3.,4.]])
print(points)
points[0][1] = 5
print(points)
```

```python
print(points.storage())
print('\n',C.storage())
print('\n',reshaped.storage())
```

```python
print(C.stride())
print('\n', reshaped.stride())
print(f'\n{points.stride()}\n\n{points.t().stride()}')
```

```python
X = torch.tensor([1,2,3,4])
print(X)
print('\n',torch.unsqueeze(X, 0))
print('\n',torch.unsqueeze(X, 1))
print('\n',torch.unsqueeze(X, 2))
```

# Representar datos con tensores

```python
print(points)
print('\n',torch.mean(points))
print('\n',torch.mean(points,dim=0))
print('\n',torch.mean(points,dim=1))
```

```python
print(points)
print('\n',torch.std(points))
print('\n',torch.std(points,dim=0))
print('\n',torch.std(points,dim=1))
```
